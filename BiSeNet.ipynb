```python
# 코랩 셀에 이 코드 붙이기
import os
import cv2
from PIL import Image, ImageOps
import subprocess

# 원본 이미지 경로
input_image_path = "/content/face-parsing/assets/images/me4.jpg"

# EXIF 회전 처리된 이미지 저장할 임시 경로
fixed_image_path = "/content/fixed_image_temp.jpg"

# 1단계: EXIF 처리
img = Image.open(input_image_path)
img = ImageOps.exif_transpose(img)  # ★ 회전 적용
img.save(fixed_image_path)

# 2단계: 처리된 이미지로 inference.py 실행
os.chdir('/content/face-parsing')
cmd = f'python inference.py --model resnet34 --weight ./weights/resnet34.pt --input {fixed_image_path} --output ./results'
subprocess.run(cmd, shell=True)

print("✅ EXIF 처리 + 추론 완료!")
```
